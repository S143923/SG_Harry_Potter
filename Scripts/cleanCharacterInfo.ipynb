```python
import json
import re
from collections import defaultdict
```

```python
def save_as_json(data, file_name):
    with open("./" + file_name + ".json", 'w') as f:
        f.write(json.dumps(data))
```

```python
with open("characterInfo.json") as json_file:
    characters = json.load(json_file)
```

```python
stop_words = {"", " ", "\n", "None"}
```

```python
for character, info in characters.iteritems():
    for key, value in info.iteritems():
        if key == "loyalty":
            string = value[0]
            if string not in stop_words:
                loyalties =  string.split("\n")
                #remove references
                characters[character][key] = [loyalty.split("<")[0].split("*")[-1] for loyalty in loyalties if loyalty not in stop_words]
            else: 
                characters[character][key] = u"none"
        else:
            if type(value) == list:
                string = value[0]
                if string not in stop_words:
                    #remove references
                    characters[character][key] = string.split("<")[0]
                else:
                    characters[character][key] = u"none"
            
```

```python
blood_dict = defaultdict(str)
```

```python
for character, info in characters.iteritems():
    blood = info["blood"]
    blood_dict[blood.lower()] = ""
```

```python
loyalty_dict = defaultdict(list)
```

```python
for character, info in characters.iteritems():
    loyalties = info["loyalty"]
    for loyalty in loyalties:
        if "[[" in loyalty:
            content = re.search(r"\[\[(.*?)\]\]", loyalty).groups()[0].lower()
            loyalty_dict[content] = content
        else:
            lower = loyalty.lower()
            loyalty_dict[lower] = lower
```

```python
# clean manually after grouping all values
#save_as_json(loyalty_dict, "loyaltyDict2")
```

```python
with open("loyaltyDict.json") as json_file:
    loyalty_dict = json.load(json_file)
```

```python
clean_character_dict = {}
```

```python
def parseLoyalty(loyalty):
    if "[[" in loyalty:
            return re.search(r"\[\[(.*?)\]\]", loyalty).groups()[0].lower()
    else:
        return loyalty.lower()
```

```python
for character, info in characters.iteritems():
    blood = blood_dict[info["blood"].lower()].strip().lower()
    house = info["house"].strip().lower()
    loyalty = [loyalty_dict[parseLoyalty(loyalty)].strip().lower() for loyalty in info["loyalty"] if loyalty_dict.has_key(parseLoyalty(loyalty))]
    clean_character_dict[character] = {"blood":blood,"house":house,"loyalty":loyalty}
    
```
